# 1. FRAMING THE PROBLEM:
### The goal of this report is to show the relationship between the median household value (dependant variable) with three independant variables namely,median income, population and number of households.
### This is a univariate linear regression model.

# 2. GETTING THE DATA
### The Data of source file' house.csv' was downloaded online with link sent to us by our Professor.

```python
import pandas as pd
import seaborn as sns

file_path = 'housing.csv'
dataframe = pd.read_csv(file_path)
```

# 3.Exploratory Data Analysis(EDA)


```python
print(dataframe.describe())
print(dataframe['median_house_value'].describe())
print(dataframe['median_income'].describe())
print(dataframe['population'].describe())
print(dataframe['households'].describe)
dataframe.head()
```

# 4. RUNNING LINEAR REGRESSIONS


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x='median_house_value', y='median_income',data=dataframe)
plt.title('median income v median house value')
plt.show()

```

```python
sns.scatterplot(x='median_house_value', y='population',data=dataframe, color= "green")
plt.title('population v median house value')
plt.show()
```

```python
sns.scatterplot(x='median_house_value', y='households',data=dataframe, color = 'indigo')
plt.title('number of households v median house value')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
x = dataframe[["median_house_value"]]  
y = dataframe["median_income"]  

# Fit the model
model1.fit(x, y)

print(f"Intercept: {model1.intercept_}")
print(f"Coefficients: {model1.coef_}")
```

```python
model2 = LinearRegression()
x = dataframe[["median_house_value"]]  
y = dataframe["population"]  

# Fit the model
model2.fit(x, y)

print(f"Intercept: {model2.intercept_}")
print(f"Coefficients: {model2.coef_}")
```

```python
model3 = LinearRegression()
x = dataframe[["median_house_value"]]  
y = dataframe["households"]  

# Fit the model
model3.fit(x, y)

print(f"Intercept: {model3.intercept_}")
print(f"Coefficients: {model3.coef_}")
```

# 6. PLOTTING LINE ON TOP OF SCATTER PLOTS

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

sns.scatterplot(x="median_house_value",y="median_income", data=dataframe, color='blue')
plt.plot(dataframe[['median_house_value']],model1.predict(dataframe[['median_house_value']]), color="black",linewidth=3)
plt.show()

print(f"mean squared error:{mean_squared_error(dataframe[['median_income']], model1.predict(dataframe[['median_house_value']]))}")
print(f"mean absolute error:{mean_absolute_error(dataframe[['median_income']], model1.predict(dataframe[['median_house_value']]))}")
```

```python
sns.scatterplot(x="median_house_value",y="population", data=dataframe, color='green')
plt.plot(dataframe[['median_house_value']],model2.predict(dataframe[['median_house_value']]), color="black",linewidth=3)
plt.show()

print(f"mean squared error:{mean_squared_error(dataframe[['population']], model2.predict(dataframe[['median_house_value']]))}")
print(f"mean absolute error:{mean_absolute_error(dataframe[['population']], model2.predict(dataframe[['median_house_value']]))}")
```

```python
sns.scatterplot(x="median_house_value",y="households", data=dataframe, color='indigo')
plt.plot(dataframe[['median_house_value']],model3.predict(dataframe[['median_house_value']]), color="black",linewidth=3)
plt.show()

print(f"mean squared error:{mean_squared_error(dataframe[['households']], model3.predict(dataframe[['median_house_value']]))}")
print(f"mean absolute error:{mean_absolute_error(dataframe[['households']], model3.predict(dataframe[['median_house_value']]))}")
```

# 7. CONCLUSION
From the above prediction models, the following deductions can be easily seen:
i: The number of households has little or no influence on the median house value
ii: The population has little or no influence on the median house value as the prediction line is flat.
iii: The median income, however has an effect on the median house value. The higer the income, the higher the house value.
